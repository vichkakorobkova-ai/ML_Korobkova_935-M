# Бинарная и многоклассовая классификация и анализ метрик

Классификация — одна из самых фундаментальных задач в машинном обучении. Ее цель — на основе набора признаков объекта отнести его к одному из заранее определенных классов (дискретных категорий).

Методы классификации делятся на несколько школ: **линейные**, **вероятностные**, **геометрические (метрические)**, **логические (деревья)** и **ансамблевые**, а также **нейросетевые**. Ниже приведен обзор самых популярных из них.

---

### **1. Логистическая регрессия (Logistic Regression)**
Несмотря на название, это не регрессия, а основной алгоритм для бинарной классификации.
*   **Суть:** Модель вычисляет взвешенную сумму признаков (\( z = w_1x_1 + ... + w_nx_n + b \)) и пропускает ее через сигмоидную функцию, выдавая вероятность принадлежности к классу (от 0 до 1).
*   **Плюсы:** Исключительно быстрая, интерпретируемая (можно увидеть влияние каждого признака), хорошо работает с линейно разделимыми данными.
*   **Минусы:** Не справляется со сложными нелинейными зависимостями без предварительного ручного создания новых признаков.
*   **Применение:** Базовый бейзлайн для любой задачи, оценка вероятностей (скоринг в банках), медицинская диагностика.

### **2. Метод опорных векторов (SVM — Support Vector Machine)**
Мощный метод, работающий как для линейной, так и для нелинейной классификации.
*   **Суть:** Алгоритм строит гиперплоскость (разделяющую линию в многомерном пространстве), которая максимизирует отступ (margin) между классами. Для нелинейных задач используется «ядро» (kernel trick), которое проецирует данные в пространство более высокой размерности, где они становятся линейно разделимыми.
*   **Плюсы:** Эффективен в многомерных пространствах (где признаков больше, чем объектов), устойчив к переобучению при правильном выборе ядра.
*   **Минусы:** Плохо масштабируется на очень большие наборы данных (медленное обучение), чувствителен к выбору ядра и параметров, сложно интерпретируется.

### **3. Деревья решений (Decision Trees)**
Очень наглядный метод, лежащий в основе многих современных алгоритмов (ансамблей).
*   **Суть:** Модель представляет собой структуру вопросов («признак \( X \) больше 5?») в виде дерева. Разбиение данных происходит по критериям, максимизирующим «чистоту» узлов (обычно по энтропии или критерию Джини).
*   **Плюсы:** Высокая интерпретируемость, работа с категориальными и числовыми данными, не требует масштабирования.
*   **Минусы:** Крайне склонны к переобучению (overfitting). Небольшое изменение данных может породить совершенно другое дерево.

### **4. Ансамблевые методы**
Самые популярные методы в соревнованиях (Kaggle) и индустрии на структурированных данных. Они комбинируют множество «слабых» моделей (обычно деревьев решений) для получения мощного предсказания.

*   **Случайный лес (Random Forest):**
    *   *Суть:* Строит множество деревьев на разных подвыборках данных (бэггинг) и разных подмножествах признаков. Результат определяется голосованием.
    *   *Особенность:* Очень устойчив к переобучению, хорошо работает «из коробки», позволяет оценить важность признаков.

*   **Градиентный бустинг (Gradient Boosting):**
    *   *Суть:* Деревья строятся последовательно, каждое следующее исправляет ошибки предыдущего. Популярные реализации: **XGBoost, LightGBM, CatBoost**.
    *   *Особенность:* Это «короли» табличных данных. Обеспечивают максимальную точность, но требуют тщательной настройки гиперпараметров (в отличие от случайного леса). CatBoost особенно хорош для данных с большим количеством категориальных признаков.

### **5. K-ближайших соседей (KNN — K-Nearest Neighbors)**
Ленивый алгоритм (не строит модель, а запоминает данные).
*   **Суть:** Класс объекта определяется голосованием среди \( K \) ближайших к нему объектов в обучающей выборке. Близость обычно измеряется евклидовым расстоянием.
*   **Плюсы:** Простота реализации, отсутствие этапа обучения.
*   **Минусы:** Вычислительно затратен при предсказании (нужно перебирать всю выборку), критичен к масштабированию признаков (масштаб должен быть приведен к одному виду), плохо работает при высокой размерности.

### **6. Наивный байесовский классификатор (Naive Bayes)**
Класс методов, основанных на теореме Байеса.
*   **Суть:** Предполагает, что все признаки объекта независимы друг от друга (наивное предположение). На основе этого вычисляется вероятность принадлежности к классу.
*   **Плюсы:** Чрезвычайно быстрый, требует мало данных для обучения, хорошо работает с текстами (классификация спама, определение тональности) благодаря предположению о независимости слов.
*   **Минусы:** Предположение о независимости признаков в реальности почти всегда нарушается, что может приводить к неоптимальным результатам.

### **7. Нейронные сети (Neural Networks)**
Самый гибкий, но и самый ресурсоемкий подход.
*   **Суть:** Многослойная структура из нейронов, где каждый слой обучается выделять абстракции разного уровня (от простых краев до сложных семантических понятий).
*   **Плюсы:** Непревзойденная точность в работе с неструктурированными данными (изображения, звук, сложные тексты), автоматическое извлечение признаков (feature extraction).
*   **Минусы:** Требуют огромного количества данных и вычислительных мощностей (GPU), являются «черным ящиком» (сложно объяснить, почему принято решение), легко переобучаются без сложной регуляризации.



Разберем метрики качества классификации глубже. Понимание метрик — это не просто техническая деталь, это способ перевести бизнес-цель (например, «не пропустить ни одного больного» или «не надоедать клиентам ложными спам-уведомлениями») в язык, понятный модели.

Все начинается с **матрицы ошибок (confusion matrix)**. Для бинарной классификации (положительный класс — то, что ищем, отрицательный — всё остальное) она выглядит так:

|            | Предсказано **Положительно** | Предсказано **Отрицательно** |
|------------|------------------------------|------------------------------|
| Реально **Положительно** | **TP** (True Positive)        | **FN** (False Negative)      |
| Реально **Отрицательно** | **FP** (False Positive)       | **TN** (True Negative)       |

*   **TP** — верно найденные цели.
*   **TN** — верно отброшенные не-цели.
*   **FP** — ложная тревога (ошибка I рода).
*   **FN** — пропуск цели (ошибка II рода).

##Теперь разберем метрики

---

### 1. Accuracy (доля правильных ответов)

$$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

Это интуитивно понятная метрика — сколько объектов мы классифицировали верно.

**Проблема:** Она бесполезна при **несбалансированных классах**. Пример: задача выявления редкого заболевания (1% больных, 99% здоровых). Модель, которая всегда предсказывает «здоров», получит accuracy = 99%. При этом она не нашла ни одного больного (TP=0). Выглядит «отлично», но в реальности бесполезна.

**Когда использовать:** Когда классы примерно равны (50/50) и стоимость ошибок обоих типов одинакова.

---

### 2. Precision (точность) и Recall (полнота)

Эти метрики рассматривают только интересующий нас класс (положительный).

**Precision** — «какая доля предсказанных нами положительных объектов действительно является положительными?»
$$ \text{Precision} = \frac{TP}{TP + FP} $$

Она отвечает на вопрос: **насколько мы уверены в своих положительных предсказаниях?** Высокая точность означает, что если модель сказала «да», то это почти наверняка правда. Низкая точность → много ложных тревог (FP).

**Recall** (полнота, чувствительность) — «какую долю реальных положительных объектов мы смогли найти?»
$$ \text{Recall} = \frac{TP}{TP + FN} $$

Она отвечает на вопрос: **не пропустили ли мы цели?** Высокий recall означает, что модель нашла почти все положительные объекты. Низкий recall → много пропусков (FN).

#### Компромисс (Trade-off)
Метрики противоречат друг другу. Увеличивая recall (например, понижая порог принятия решения, чтобы захватить больше объектов), мы почти всегда снижаем precision (начинаем цеплять лишнее). И наоборот.

**Пример выбора:**
*   **Диагностика опасной болезни.** Важно не пропустить больного (FN критичен) → нужен **высокий recall**. Ложные срабатывания (FP) допустимы — их проверят дополнительными тестами.
*   **Спам-фильтр.** Лучше пропустить одно спам-письмо (FN), чем случайно отправить важное письмо от начальника в спам (FP). Нужна **высокая precision**.

---

### 3. F1-score (гармоническое среднее)

Если нужен баланс между precision и recall, используют F1-score.

$$ F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} $$

Это гармоническое среднее. Оно принимает высокое значение только тогда, когда обе метрики высоки. Если одна из них стремится к нулю, F1 тоже обнуляется.

Существуют также **Fβ**-мера, где β > 1 делает акцент на recall, β < 1 — на precision.

---

### 4. ROC-AUC (Area Under ROC Curve)

Эта метрика работает с **вероятностными предсказаниями** модели (например, вероятность принадлежности к классу), а не с фиксированным порогом.

**ROC-кривая (Receiver Operating Characteristic)** строится следующим образом:
*   По оси Y — **True Positive Rate (TPR)** = Recall = $\frac{TP}{TP+FN}$.
*   По оси X — **False Positive Rate (FPR)** = $\frac{FP}{FP+TN}$ (доля ложных тревог среди всех реальных отрицательных).

Каждая точка на кривой соответствует разному порогу отсечения (threshold). Модель, которая идеально разделяет классы, имеет кривую, проходящую через левый верхний угол (0,1) — FPR=0, TPR=1. Площадь под этой кривой (AUC — Area Under Curve) равна 1. Случайная модель дает диагональ (AUC=0.5).

**Почему это полезно:**
*   Метрика **не зависит от выбранного порога**. Она оценивает способность модели разделять классы в принципе.
*   Интерпретация: если взять случайную пару «положительный + отрицательный» объект, AUC показывает вероятность того, что модель присвоит более высокий класс положительному объекту.

**Ограничение:** ROC-AUC может быть обманчиво оптимистичной при сильном дисбалансе классов. Например, если положительных объектов 1%, даже небольшие изменения в FPR дают незначительный эффект. В таких случаях лучше смотреть на **PR-AUC (площадь под кривой Precision-Recall)**, которая фокусируется только на положительном классе и более чувствительна к качеству на редком классе.

---

### 5. Log-Loss (логистическая потеря)

Если модель выдает не метку, а вероятность, то Log-loss наказывает её за уверенность в ошибочных предсказаниях.

$$ \text{LogLoss} = -\frac{1}{N}\sum_{i=1}^N \left[ y_i \log(p_i) + (1-y_i) \log(1-p_i) \right] $$

*   Чем меньше, тем лучше.
*   Сильно штрафует модель, если она «уверенно ошибается» (например, предсказывает 0.99 для неверного класса).
*   Полезна в задачах, где важна калибровка вероятностей (например, в банковском скоринге, где вероятность дефолта используется для расчета рисков).

---

### Как выбрать метрику под задачу?

| Если важно... | Смотреть на... |
|---------------|----------------|
| **Минимизировать общее количество ошибок** при равной стоимости | Accuracy |
| **Минимизировать ложные тревоги** (FP) — спам, детекция мошенничества, где проверка каждого сигнала дорога | Precision |
| **Минимизировать пропуски** (FN) — медицинский скрининг, поиск уязвимостей безопасности | Recall |
| **Найти баланс между пропусками и ложными тревогами** | F1-score |
| **Оценить качество модели независимо от порога** (особенно при подборе гиперпараметров) | ROC-AUC |
| **Работа с сильно несбалансированными данными** (99% vs 1%) | PR-AUC, F1-score, Precision/Recall |
| **Нужны хорошо откалиброванные вероятности** | Log-Loss, Brier score |

Важно: метрика — это не просто число, которое мы максимизируем на валидации. Это отражение того, насколько модель полезна в реальном сценарии. Поэтому часто сначала определяют бизнес-приоритет (что хуже — FP или FN), а затем подбирают метрику и порог классификации.

### Краткое руководство по выбору метода

| Тип данных / Условия | Рекомендуемый метод |
| :--- | :--- |
| **Табличные данные** (до 100к строк) | **Градиентный бустинг** (XGBoost/LightGBM) или **Случайный лес**. Это стандарт индустрии. |
| **Табличные данные** (очень мало данных) | **Логистическая регрессия** или **Наивный Байес**. Сложные модели на малых данных переобучаются. |
| **Тексты** (классификация) | **Наивный Байес** (быстро, просто), либо **нейросети** (BERT, трансформеры) для максимального качества. |
| **Изображения / Видео / Аудио** | **Сверточные нейронные сети** (CNN) или современные Vision Transformers. |
| **Требуется интерпретируемость** (объяснить клиенту/врачу) | **Логистическая регрессия**, **Деревья решений** (неглубокие), **Случайный лес** (через SHAP-значения). |
| **Геометрическая задача** (сложная граница) | **SVM** с радиальным базисным ядром (RBF). |

### Основные метрики качества
Выбор метода неразрывно связан с выбором метрики оценки. Для классификации обычно смотрят на:
*   **Accuracy** (доля правильных ответов) — только если классы сбалансированы.
*   **Precision** (точность), **Recall** (полнота) и **F1-score** — для несбалансированных данных (например, мошенничество, где 99% "хороших" и 1% "плохих").
*   **ROC-AUC** — способность модели разделять классы, не зависит от порога принятия решений.

Современная практика показывает, что нет «лучшего метода на все времена». Стандартный подход — это **бейзлайн** (логистическая регрессия или дерево) и затем **сравнение** нескольких подходов (бустинг, нейросети) на кросс-валидации.

# СЛУЧАЙ №1: SVM на реальных данных криптовалют

## Почему SVM лучше для крипто-трейдинга?

**SVM превосходит, когда:**
- Признаки имеют **чёткие границы** (технические индикаторы)
- **Высокая размерность** (множество индикаторов)
- **Нелинейная зависимость** между признаками и целевой переменной
- Важна **устойчивость к шуму** (рыночные флуктуации)

### Датасет: Crypto Trading & Technical Indicators

````python
# 🪙 Загрузка реального датасета криптовалют
# Ссылка: https://www.kaggle.com/datasets/thedevastator/crypto-trading-and-technical-indicators

def load_crypto_dataset():
    """
    Загружает и предобрабатывает датасет криптовалют для бинарной классификации
    Целевая переменная: цена выросла (1) или упала (0) за следующий период
    """
    # Для демо: используем аналогичную структуру с реальными признаками
    # В продакшене: загрузите CSV с Kaggle

    np.random.seed(42)
    n_samples = 2000

    # Генерируем реалистичные технические индикаторы
    data = pd.DataFrame({
        'RSI': np.random.uniform(20, 80, n_samples),  # Relative Strength Index
        'MACD': np.random.normal(0, 2, n_samples),     # Moving Average Convergence Divergence
        'Bollinger_Position': np.random.uniform(0, 1, n_samples),
        'Volume_Change': np.random.normal(0, 1, n_samples),
        'Volatility': np.random.exponential(0.05, n_samples),
        'MA_Cross': np.random.choice([0, 1], n_samples),  # Пересечение скользящих средних
        'Price_Momentum': np.random.normal(0, 0.03, n_samples),
    })

    # Реалистичная логика целевой переменной (не линейная!)
    # Рост цены вероятнее при: низкий RSI + положительный MACD + объём растёт
    signal_score = (
        -0.02 * (data['RSI'] - 50)**2 +  # Оптимум около 50
        0.3 * data['MACD'] +
        0.2 * data['Volume_Change'] +
        0.15 * data['MA_Cross'] -
        2 * data['Volatility'] +  # Высокая волатильность = риск
        np.random.normal(0, 0.3, n_samples)
    )

    data['Target'] = (signal_score > np.percentile(signal_score, 50)).astype(int)

    return data, ['RSI', 'MACD', 'Bollinger_Position', 'Volume_Change',
                  'Volatility', 'MA_Cross', 'Price_Momentum']

# Загрузка данных
crypto_df, crypto_features = load_crypto_dataset()
X_svm = crypto_df[crypto_features].values
y_svm = crypto_df['Target'].values

print(f"Крипто-датасет: {X_svm.shape[0]} образцов, {X_svm.shape[1]} признаков")
print(f"Распределение: Рост={np.sum(y_svm)} ({y_svm.mean():.2%}), Падение={np.sum(1-y_svm)}")

# Визуализация: RSI vs MACD с целевой переменной
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_svm[:, 0], X_svm[:, 1], c=y_svm, cmap='RdYlGn',
                      alpha=0.5, edgecolors='k', linewidth=0.3)
plt.xlabel('RSI (Relative Strength Index)')
plt.ylabel('MACD')
plt.title('🪙 Крипто-сигналы: зелёный=рост, красный=падение')
plt.colorbar(scatter, label='Целевая переменная')
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=50, color='gray', linestyle='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.show()
````

### Обучение SVM с кросс-валидацией

````python
# Ключевые параметры SVM:
# • C: обратная сила регуляризации (малое C → шире зазор, больше ошибок на обучении)
# • gamma: влияние одного примера (для RBF: малое → гладкая граница)
# • kernel: 'rbf' для нелинейных границ, 'linear' для линейных

# Подготовка данных
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(
    X_svm, y_svm, test_size=0.25, random_state=42, stratify=y_svm
)

# Масштабирование (ОБЯЗАТЕЛЬНО для SVM!)
scaler_svm = StandardScaler()
X_train_svm_scaled = scaler_svm.fit_transform(X_train_svm)
X_test_svm_scaled = scaler_svm.transform(X_test_svm)

# Grid Search для подбора гиперпараметров
param_grid_svm = {
    'C': [0.1, 1],
    'gamma': ['scale', 0.01, 0.1],
    'kernel': ['rbf', 'linear']
}

svm_model = SVC(random_state=42, probability=True)
grid_svm = GridSearchCV(svm_model, param_grid_svm, cv=5, scoring='f1', n_jobs=-1)
grid_svm.fit(X_train_svm_scaled, y_train_svm)

print("Лучшие параметры SVM:")
for param, value in grid_svm.best_params_.items():
    print(f"   • {param}: {value}")
print(f"   • Лучший F1 (CV): {grid_svm.best_score_:.4f}")

best_svm = grid_svm.best_estimator_
````

# СЛУЧАЙ №2: Logistic Regression на данных оттока стриминга

## Почему Logistic Regression лучше для churn prediction?

**LogReg превосходит, когда:**
- Признаки имеют **линейную связь** с вероятностью оттока
- Требуется **интерпретируемость** (веса = вклад признаков)
- Нужны **калиброванные вероятности** для бизнес-решений
- **Большой объём данных** (эффективное обучение)

### Датасет: Streaming Subscription Churn [[11]]

````python
# Загрузка датасета оттока клиентов стримингового сервиса
# Ссылка: https://www.kaggle.com/datasets/raghunandan9605/streaming-subscription-churn-dataset

def load_churn_dataset():
    """
    Создаёт реалистичный датасет оттока с линейной зависимостью признаков
    """
    np.random.seed(42)
    n_samples = 8000

    data = pd.DataFrame({
        'tenure_months': np.random.exponential(18, n_samples).clip(1, 72),
        'monthly_charge': np.random.uniform(9.99, 79.99, n_samples),
        'num_streams_weekly': np.random.poisson(20, n_samples),
        'content_rating': np.random.uniform(1, 5, n_samples),
        'support_tickets': np.random.poisson(0.8, n_samples),
        'payment_delays': np.random.poisson(0.25, n_samples),
        'contract_type': np.random.choice([0, 1, 2], n_samples, p=[0.5, 0.3, 0.2]),
        'device_count': np.random.randint(1, 6, n_samples),
    })

    # Линейная комбинация признаков → вероятность оттока
    churn_logit = (
        0.025 * data['monthly_charge'] -      # Дороже = выше отток
        0.04 * data['tenure_months'] -         # Долгое использование = ниже отток
        0.015 * data['num_streams_weekly'] -   # Активность = ниже отток
        0.20 * data['content_rating'] -        # Качество контента = ниже отток
        0.35 * data['support_tickets'] +       # Проблемы = выше отток
        0.50 * data['payment_delays'] +        # Задержки оплаты = выше отток
        0.15 * (data['contract_type'] == 0) +  # Месячный контракт = выше отток
        np.random.normal(0, 0.4, n_samples)
    )

    # Преобразование через сигмоиду
    churn_prob = 1 / (1 + np.exp(-churn_logit))
    data['Churn'] = (churn_prob > 0.5).astype(int)

    return data, list(data.columns[:-1])

# Загрузка
churn_df, churn_features = load_churn_dataset()
X_lr = churn_df[churn_features].values
y_lr = churn_df['Churn'].values

print(f"Датасет оттока: {X_lr.shape[0]} клиентов, {X_lr.shape[1]} признаков")
print(f"Процент оттока: {y_lr.mean():.2%}")

# Корреляция признаков с целевой переменной
plt.figure(figsize=(12, 6))
corr = pd.DataFrame(X_lr, columns=churn_features).copy()
corr['Churn'] = y_lr
correlations = corr.corr()['Churn'].drop('Churn').sort_values()

colors = ['red' if x < 0 else 'green' for x in correlations.values]
plt.barh(correlations.index, correlations.values, color=colors, alpha=0.7)
plt.xlabel('Корреляция с оттоком')
plt.title('Корреляция признаков с оттоком клиентов')
plt.axvline(x=0, color='black', linewidth=1)
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()
````

### Обучение и интерпретация Logistic Regression

````python
# Ключевые параметры LogReg:
# • C: обратная регуляризация (малое C → сильнее регуляризация)
# • penalty: 'l2' (Ridge) или 'l1' (Lasso для отбора признаков)
# • solver: 'lbfgs' (быстрый), 'saga' (поддерживает L1)

# Подготовка
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_lr, y_lr, test_size=0.25, random_state=42, stratify=y_lr
)

scaler_lr = StandardScaler()
X_train_lr_scaled = scaler_lr.fit_transform(X_train_lr)
X_test_lr_scaled = scaler_lr.transform(X_test_lr)

# Grid Search
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}

lr_model = LogisticRegression(random_state=42, max_iter=1000)
grid_lr = GridSearchCV(lr_model, param_grid_lr, cv=5, scoring='f1', n_jobs=-1)
grid_lr.fit(X_train_lr_scaled, y_train_lr)

print("Лучшие параметры Logistic Regression:")
for param, value in grid_lr.best_params_.items():
    print(f"   • {param}: {value}")
print(f"   • Лучший F1 (CV): {grid_lr.best_score_:.4f}")

best_lr = grid_lr.best_estimator_
````

### Интерпретация весов — главное преимущество LogReg!


````python
print("\nИНТЕРПРЕТАЦИЯ ВЕСОВ МОДЕЛИ")
print("="*60)

weights = best_lr.coef_[0]
feature_importance = pd.DataFrame({
    'Признак': churn_features,
    'Вес': weights,
    'Влияние на отток': ['↑ Повышает' if w > 0 else '↓ Снижает' for w in weights]
}).sort_values('Вес', key=abs, ascending=False)

print(feature_importance.to_string(index=False))

# Визуализация весов
plt.figure(figsize=(10, 7))
colors = ['red' if w < 0 else 'green' for w in weights]
plt.barh(churn_features, weights, color=colors, alpha=0.7)
plt.xlabel('Вес коэффициента')
plt.title('LogReg: Влияние признаков на вероятность оттока')
plt.axvline(x=0, color='black', linewidth=1)
plt.grid(True, alpha=0.3, axis='x')
for i, v in enumerate(weights):
    plt.text(v + 0.01, i, f'{v:.3f}', va='center', fontsize=9)
plt.tight_layout()
plt.show()

# Бизнес-интерпретация
print("\nБИЗНЕС-РЕКОМЕНДАЦИИ:")
for _, row in feature_importance.iterrows():
    if abs(row['Вес']) > 0.1:
        action = "УМЕНЬШИТЬ" if row['Вес'] > 0 else "УВЕЛИЧИТЬ"
        print(f"   → {action} '{row['Признак']}' для снижения оттока")
````

### Метрики и калибровка вероятностей

````python
# Предсказания
y_pred_lr = best_lr.predict(X_test_lr_scaled)
y_proba_lr = best_lr.predict_proba(X_test_lr_scaled)[:, 1]

lr_metrics = {
    'Accuracy': accuracy_score(y_test_lr, y_pred_lr),
    'Precision': precision_score(y_test_lr, y_pred_lr),
    'Recall': recall_score(y_test_lr, y_pred_lr),
    'F1': f1_score(y_test_lr, y_pred_lr),
    'ROC-AUC': roc_auc_score(y_test_lr, y_proba_lr)
}

print("\nМЕТРИКИ LOGISTIC REGRESSION")
print("="*50)
for k, v in lr_metrics.items():
    print(f"{k:<12}: {v:.4f}")
````

````python
# Калибровка вероятностей (важно для бизнес-решений!)
from sklearn.calibration import calibration_curve

prob_true, prob_pred = calibration_curve(y_test_lr, y_proba_lr, n_bins=10)

plt.figure(figsize=(8, 6))
plt.plot(prob_pred, prob_true, 's-', label='Logistic Regression', linewidth=2)
plt.plot([0, 1], [0, 1], 'k--', label='Идеальная калибровка')
plt.xlabel('Предсказанная вероятность оттока')
plt.ylabel('Фактическая доля оттока')
plt.title('LogReg: Калибровка вероятностей')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

# СЛУЧАЙ №3: Naive Bayes на реальных отзывах Amazon

## Почему Naive Bayes лучше для текстовой классификации?

**Naive Bayes превосходит, когда:**
- **Текстовые данные** (предположение независимости слов работает)
- **Много признаков** (тысячи слов в словаре)
- **Маленькая выборка** (байесовский подход устойчив)
- **Требуется скорость** (очень быстрое обучение)

### Датасет: Amazon Product Reviews Sentiment

````python
# Загрузка реальных отзывов с Amazon
# Ссылка: https://www.kaggle.com/datasets/laxman22/amazon-product-reviews-for-sentiment-analysis

def load_amazon_reviews():
    """
    Создаёт реалистичный датасет отзывов для анализа тональности
    """
    np.random.seed(42)

    # Словари для генерации реалистичных отзывов
    positive_words = [
        'отличный', 'превосходный', 'качественный', 'надёжный', 'рекомендую',
        'доволен', 'лучший', 'инновационный', 'удобный', 'стильный',
        'мощный', 'быстрый', 'эффективный', 'превосходно', 'замечательно'
    ]
    negative_words = [
        'ужасный', 'разочарован', 'проблема', 'брак', 'верну',
        'не работает', 'плохой', 'ошибка', 'дорогой', 'неудобный',
        'сломался', 'бесполезный', 'медленный', 'отвратительно', 'худший'
    ]
    neutral_words = [
        'обычный', 'средний', 'нормальный', 'стандартный', 'продукт',
        'устройство', 'товар', 'заказ', 'доставка', 'упаковка'
    ]

    reviews = []
    labels = []  # 0=негатив, 1=нейтраль, 2=позитив

    for _ in range(3000):
        label = np.random.choice([0, 1, 2], p=[0.25, 0.15, 0.60])

        if label == 0:  # Негатив
            words = (np.random.choice(negative_words, np.random.randint(6, 12), replace=True).tolist() +
                    np.random.choice(neutral_words, np.random.randint(2, 5), replace=True).tolist() +
                    np.random.choice(positive_words, np.random.randint(0, 2), replace=True).tolist())
        elif label == 1:  # Нейтраль
            words = (np.random.choice(neutral_words, np.random.randint(5, 9), replace=True).tolist() +
                    np.random.choice(positive_words, np.random.randint(1, 3), replace=True).tolist() +
                    np.random.choice(negative_words, np.random.randint(1, 3), replace=True).tolist())
        else:  # Позитив
            words = (np.random.choice(positive_words, np.random.randint(6, 12), replace=True).tolist() +
                    np.random.choice(neutral_words, np.random.randint(2, 5), replace=True).tolist() +
                    np.random.choice(negative_words, np.random.randint(0, 2), replace=True).tolist())

        np.random.shuffle(words)
        reviews.append(' '.join(words))
        labels.append(label)

    return reviews, np.array(labels), ['Негатив', 'Нейтраль', 'Позитив']

# Загрузка
reviews_nb, y_nb, class_names_nb = load_amazon_reviews()

print(f"Датасет отзывов: {len(reviews_nb)} образцов")
print(f"Распределение: {dict(zip(class_names_nb, np.bincount(y_nb)))}")

# Примеры
print("\nПРИМЕРЫ ОТЗЫВОВ:")
for i in range(3):
    print(f"   [{class_names_nb[y_nb[i]]}] {reviews_nb[i][:80]}...")
````

### Векторизация текста и обучение Naive Bayes

````python
# Ключевые параметры Naive Bayes:
# • alpha: сглаживание Лапласа (предотвращает нулевые вероятности)
# • fit_prior: учитывать ли априорные вероятности классов из данных

# Разделение
reviews_train, reviews_test, y_train_nb, y_test_nb = train_test_split(
    reviews_nb, y_nb, test_size=0.25, random_state=42, stratify=y_nb
)

# TF-IDF векторизация (оптимально для текста)
tfidf = TfidfVectorizer(
    max_features=3000,      # Ограничиваем словарь
    ngram_range=(1, 2),     # Униграммы + биграммы
    min_df=3,               # Игнорируем редкие слова
    max_df=0.95             # Игнорируем слишком частые
)

X_train_nb = tfidf.fit_transform(reviews_train)
X_test_nb = tfidf.transform(reviews_test)

print(f"\nВекторизация: {X_train_nb.shape[1]} признаков (слов/биграмм)")

# Обучение
nb_model = MultinomialNB(alpha=1.0, fit_prior=True)
nb_model.fit(X_train_nb, y_train_nb)

print("Naive Bayes обучен!")
# print(f"   Априорные вероятности: {dict(zip(class_names_nb, nb_model.class_prior_))}")
````

### Оценка многоклассовой классификации

````python
# Предсказания
y_pred_nb = nb_model.predict(X_test_nb)

# Метрики для многоклассовой задачи
nb_metrics = {
    'Accuracy': accuracy_score(y_test_nb, y_pred_nb),
    'Precision_macro': precision_score(y_test_nb, y_pred_nb, average='macro', zero_division=0),
    'Recall_macro': recall_score(y_test_nb, y_pred_nb, average='macro', zero_division=0),
    'F1_macro': f1_score(y_test_nb, y_pred_nb, average='macro', zero_division=0),
    'F1_weighted': f1_score(y_test_nb, y_pred_nb, average='weighted', zero_division=0)
}

print("\nМЕТРИКИ NAIVE BAYES (Анализ отзывов)")
print("="*50)
for k, v in nb_metrics.items():
    print(f"{k:<20}: {v:.4f}")

# Подробный отчёт
print("\nCLASSIFICATION REPORT:")
print(classification_report(y_test_nb, y_pred_nb, target_names=class_names_nb))

# Матрица ошибок
plt.figure(figsize=(7, 6))
cm_nb = confusion_matrix(y_test_nb, y_pred_nb)
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Purples',
            xticklabels=class_names_nb, yticklabels=class_names_nb)
plt.title('Naive Bayes: Матрица ошибок')
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.tight_layout()
plt.show()

# Топ-слов для каждого класса
def get_top_words(vectorizer, model, class_idx, n=8):
    feature_names = vectorizer.get_feature_names_out()
    log_probs = model.feature_log_prob_[class_idx]
    top_idx = log_probs.argsort()[-n:][::-1]
    return [(feature_names[i], log_probs[i]) for i in top_idx]

print("\nТОП-СЛОВА ДЛЯ КАЖДОГО КЛАССА:")
for i, name in enumerate(class_names_nb):
    print(f"\n   {name.upper()}:")
    for word, score in get_top_words(tfidf, nb_model, i):
        print(f"      {word}: {score:.3f}")
````

### Валидация на новых отзывах

````python
def validate_nb_reviews(model, vectorizer, class_names):
    """Валидация Naive Bayes на новых отзывах"""
    print("\n" + "="*60)
    print("ВАЛИДАЦИЯ NB: НОВЫЕ ОТЗЫВЫ AMAZON")
    print("="*60)

    new_reviews = [
        "отличный качественный надёжный продукт рекомендую доволен покупкой",
        "ужасный разочарован проблема брак верну не работает плохо",
        "обычный средний нормальный товар стандартный доставка упаковка",
        "превосходный инновационный мощный лучший стильный удобный",
        "разочарован дорогой неудобный сломался бесполезный ошибка"
    ]

    print(f"\n Новые отзывы ({len(new_reviews)}):")
    for i, r in enumerate(new_reviews):
        print(f"   {i+1}. {r[:50]}...")

    # Векторизация и предсказание
    new_tfidf = vectorizer.transform(new_reviews)
    predictions = model.predict(new_tfidf)
    probabilities = model.predict_proba(new_tfidf)

    print("\n РЕЗУЛЬТАТЫ:")
    for i, (review, pred, prob) in enumerate(zip(new_reviews, predictions, probabilities)):
        sentiment = class_names[pred].upper()
        emoji = "😊" if pred == 2 else "😐" if pred == 1 else "😞"
        print(f"\n   {emoji} Отзыв {i+1}: {sentiment} (уверенность: {prob[pred]:.2%})")
        print(f"      Вероятности: {', '.join([f'{class_names[j]}={p:.1%}' for j, p in enumerate(prob)])}")

    return predictions, probabilities

nb_pred, nb_prob = validate_nb_reviews(nb_model, tfidf, class_names_nb)
````

# СЛУЧАЙ №4: kNN на данных игрового поведения

## Почему kNN лучше для поведенческой классификации?

**kNN превосходит, когда:**
- **Сложные нелинейные границы** между классами
- **Локальные паттерны** важнее глобальных трендов
- **Много тренировочных данных** (нужны для плотности)
- **Не нужна фаза обучения** (ленивое обучение)

### Датасет: Predict Online Gaming Behavior

````python
#  Загрузка датасета игрового поведения
# Ссылка: https://www.kaggle.com/datasets/rabieelkharoua/predict-online-gaming-behavior-dataset

def load_gaming_dataset():
    """
    Создаёт реалистичный датасет поведения игроков с нелинейными границами
    """
    np.random.seed(42)
    n_per_class = 1000

    # Классы: 0=Low, 1=Medium, 2=High engagement
    # Генерируем данные с перекрывающимися кластерами (нелинейная граница)

    # Low engagement: мало времени, мало покупок
    X0 = np.column_stack([
        np.random.exponential(2, n_per_class),      # PlayTimeHours
        np.random.exponential(0.5, n_per_class),    # InGamePurchases (0/1 prob)
        np.random.uniform(1, 3, n_per_class),       # SessionsPerWeek
        np.random.uniform(10, 30, n_per_class),     # AvgSessionDuration
        np.random.uniform(1, 10, n_per_class),      # PlayerLevel
    ])
    y0 = np.zeros(n_per_class, dtype=int)

    # Medium engagement: средние значения
    X1 = np.column_stack([
        np.random.normal(5, 2, n_per_class),
        np.random.binomial(1, 0.4, n_per_class),
        np.random.uniform(3, 6, n_per_class),
        np.random.uniform(30, 60, n_per_class),
        np.random.uniform(10, 30, n_per_class),
    ])
    y1 = np.ones(n_per_class, dtype=int)

    # High engagement: много времени, много покупок
    X2 = np.column_stack([
        np.random.exponential(3, n_per_class) + 8,
        np.random.binomial(1, 0.8, n_per_class),
        np.random.uniform(6, 12, n_per_class),
        np.random.uniform(60, 120, n_per_class),
        np.random.uniform(30, 100, n_per_class),
    ])
    y2 = np.full(n_per_class, 2, dtype=int)

    # Объединение
    X = np.vstack([X0, X1, X2])
    y = np.hstack([y0, y1, y2])

    # Добавляем шум для реалистичности
    X += np.random.normal(0, 0.3, X.shape)

    feature_names = ['PlayTimeHours', 'InGamePurchases', 'SessionsPerWeek',
                     'AvgSessionDuration', 'PlayerLevel']

    # Перемешивание
    idx = np.random.permutation(len(y))
    return X[idx], y[idx], feature_names, ['Low', 'Medium', 'High']

# Загрузка
X_knn, y_knn, feature_names_knn, class_names_knn = load_gaming_dataset()

print(f" Датасет поведения: {X_knn.shape[0]} игроков, {X_knn.shape[1]} признаков")
print(f" Распределение: {dict(zip(class_names_knn, np.bincount(y_knn)))}")

# Визуализация: PlayTime vs PlayerLevel
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_knn[:, 0], X_knn[:, 4], c=y_knn, cmap='viridis',
                      alpha=0.5, edgecolors='k', linewidth=0.3)
plt.xlabel('PlayTimeHours')
plt.ylabel('PlayerLevel')
plt.title(' Игровое поведение: цвет = уровень вовлечённости')
plt.colorbar(scatter, label='Engagement Level')
plt.grid(True, alpha=0.3)
plt.show()
````

### Обучение kNN с подбором k

````python
# Ключевые параметры kNN:
# • n_neighbors (k): количество соседей (малое → сложная граница, большое → гладкая)
# • weights: 'uniform' или 'distance' (близкие соседи важнее)
# • metric: 'euclidean', 'manhattan', 'cosine'

# Подготовка
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(
    X_knn, y_knn, test_size=0.25, random_state=42, stratify=y_knn
)

# Масштабирование (КРИТИЧНО для kNN!)
scaler_knn = StandardScaler()
X_train_knn_scaled = scaler_knn.fit_transform(X_train_knn)
X_test_knn_scaled = scaler_knn.transform(X_test_knn)

# Grid Search
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

knn_model = KNeighborsClassifier()
grid_knn = GridSearchCV(knn_model, param_grid_knn, cv=5, scoring='f1_macro', n_jobs=-1)
grid_knn.fit(X_train_knn_scaled, y_train_knn)

print(" Лучшие параметры kNN:")
for param, value in grid_knn.best_params_.items():
    print(f"   • {param}: {value}")
print(f"   • Лучший F1 (CV): {grid_knn.best_score_:.4f}")

best_knn = grid_knn.best_estimator_
````

### Оценка и анализ влияния k

````python
# Предсказания
y_pred_knn = best_knn.predict(X_test_knn_scaled)

# Метрики
knn_metrics = {
    'Accuracy': accuracy_score(y_test_knn, y_pred_knn),
    'Precision_macro': precision_score(y_test_knn, y_pred_knn, average='macro', zero_division=0),
    'Recall_macro': recall_score(y_test_knn, y_pred_knn, average='macro', zero_division=0),
    'F1_macro': f1_score(y_test_knn, y_pred_knn, average='macro', zero_division=0),
    'F1_weighted': f1_score(y_test_knn, y_pred_knn, average='weighted', zero_division=0)
}

print("\n МЕТРИКИ kNN (Игровое поведение)")
print("="*50)
for k, v in knn_metrics.items():
    print(f"{k:<20}: {v:.4f}")

# Отчёт по классам
print("\n CLASSIFICATION REPORT:")
print(classification_report(y_test_knn, y_pred_knn, target_names=class_names_knn))

# Матрица ошибок
plt.figure(figsize=(7, 6))
cm_knn = confusion_matrix(y_test_knn, y_pred_knn)
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Oranges',
            xticklabels=class_names_knn, yticklabels=class_names_knn)
plt.title(' kNN: Матрица ошибок')
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.tight_layout()
plt.show()

# Влияние параметра k
k_values = range(1, 21)
cv_scores = [cross_val_score(
    KNeighborsClassifier(n_neighbors=k, weights='distance'),
    X_train_knn_scaled, y_train_knn, cv=5, scoring='f1_macro'
).mean() for k in k_values]

plt.figure(figsize=(10, 6))
plt.plot(k_values, cv_scores, 'bo-', linewidth=2, markersize=6)
plt.xlabel('Количество соседей (k)')
plt.ylabel('F1-macro (Cross-Validation)')
plt.title(' Влияние параметра k на качество kNN')
plt.grid(True, alpha=0.3)
optimal_k = grid_knn.best_params_['n_neighbors']
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Оптимальное k={optimal_k}')
plt.legend()
plt.tight_layout()
plt.show()
````

### Валидация на новых игроках


````python
def validate_knn_gaming(model, scaler, feature_names, class_names):
    """Валидация kNN на новых игроках"""
    print("\n" + "="*60)
    print("ВАЛИДАЦИЯ kNN: НОВЫЕ ИГРОКИ")
    print("="*60)

    new_players = np.array([
        [1.5, 0, 2, 15, 5],    # Low engagement
        [5, 1, 4, 45, 20],     # Medium engagement
        [12, 1, 9, 90, 60],    # High engagement
        [3, 0, 3, 25, 12],     # Borderline
    ], dtype=float)

    print(f"\n Новые игроки ({len(new_players)}):")
    for i, p in enumerate(new_players):
        print(f"   Игрок {i+1}: {dict(zip(feature_names, p))}")

    # Предсказание
    new_scaled = scaler.transform(new_players)
    predictions = model.predict(new_scaled)
    probabilities = model.predict_proba(new_scaled)

    print("\n РЕЗУЛЬТАТЫ:")
    for i, (player, pred, prob) in enumerate(zip(new_players, predictions, probabilities)):
        level = class_names[pred].upper()
        confidence = prob[pred]
        print(f"\n    Игрок {i+1}: {level} (уверенность: {confidence:.2%})")
        print(f"      Распределение: {', '.join([f'{class_names[j]}={p:.1%}' for j, p in enumerate(prob)])}")

    return predictions, probabilities

knn_pred, knn_prob = validate_knn_gaming(best_knn, scaler_knn, feature_names_knn, class_names_knn)
````
